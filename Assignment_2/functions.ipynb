```python
import numpy as np
import matplotlib.pyplot as plt
```

### Define the X values between [-10, 10, 0.1]

```python
X = np.arange(-10, 10, 0.1)
```

### 1st function

```python
w = 5
b = 3
Y = w*X + b

plt.plot(X, Y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Y = 5X + 3')
plt.show()
```

### 2nd function

```python
Y = X**2

plt.plot(X, Y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Y = X^2')
plt.show()
```

### 3rd function

```python
Y = 1/(1+np.exp(-X))

plt.plot(X, Y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Y = 1/(1+e^(-X))')
plt.show()
```

### 4th function

```python
Y = (np.exp(X) - np.exp(-X)) / (np.exp(X) + np.exp(-X))

plt.plot(X, Y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Y = (e^X - e^(-X)) / (e^X + e^(-X))')
plt.show()
```

### 5th function

```python
w = 5
b = 3
u = fx = w*X + b
gx = 1/(1+np.exp(-u))

plt.plot(X, gx)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Y = 1/(1+e^(-u))')
plt.show()
```

### 6th function

```python
w = 5
b = 3
u = fx = w*X + b
gx = (np.exp(u) - np.exp(-u)) / (np.exp(u) + np.exp(-u))

plt.plot(X, gx)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Y = (e^u - e^(-u)) / (e^u + e^(-u))')
plt.show()
```

### 7th function

```python
w1 = 10
b1 = -3
u1 = fx = w1*X + b1
gx1 = 1/(1+np.exp(-u1))

plt.plot(X, gx1)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Y1 = 1/(1+e^(-u1))')
plt.show()
```

```python
w2 = 20
b2 = 7
u2 = fx = w2*X + b2
gx2 = 1/(1+np.exp(-u2))

plt.plot(X, gx2)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Y2 = 1/(1+e^(-u2))')
plt.show()
```

```python
w3 = 10
w4 = 5
b = -3
u3 = fx = w3 * gx1 + w4 * gx2 + b
gx3 = 1/(1+np.exp(-u3))

plt.plot(X, gx3)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Y = 1/(1+e^(-(w3*Y1 + w4*Y2 + b)))')
plt.show()
```
